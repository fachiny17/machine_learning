## Matplotlib example workflow

```python
# Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height)

# Plot data
ax.plot(x, y)

# Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# Save and show
fig.savefig("../matplotlib/sample_plot.png")
```

## Making Figures with NumPy arrays


```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make a scatter

fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Create another scatter
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="My Nut Butter Store",
       ylabel="Price ($)");
```

```python
# To present a horizontal bar
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

## Which one should be used? (pyplot or matplotlib OO method?)
1. When plotting something quickly, use pyplot method
2. When plotting something more advanced, use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(x="age", y="chol", kind="scatter", c="target");
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
# ax.set_xlim([45, 100]);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
heart_disease
```

```python
# Ploting a scatter of Heart Disease(Age) and Max Heart Rate(thalach)

fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=heart_disease["age"],
                     y=heart_disease["thalach"],
                     c=heart_disease["target"]);

# Customize the Scatter Plot
ax.set(title="Heart Disease: Age and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate(thalach)");

# Add legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(heart_disease["thalach"].mean(),
           linestyle="--");

# Add a vertical line
ax.axvline(heart_disease["age"].mean(),
           linestyle="-.",
           color="red");
```

```python
over_50
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Level",
        ylabel="Cholesterol")

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a meanline to the horizontal
ax0.axhline(over_50["chol"].mean(), linestyle="--");

# AGE AND CHOLESTEROL
# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])
# Customize ax1
ax1.set(title="Heart Disease: Age and Max Heart Rate(Thalach)",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target");

# Add mean line to the horizontal and vertical axes
ax1.axhline(over_50["thalach"].mean(), linestyle="--")
ax1.axvline(over_50["age"].mean(), linestyle="--", color="orange")

# Add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

## Customizing Matplotlib plots and gettin stylish

```python
# See the diferent styles avaliable
plt.style.available
```

```python
plt.style.use("seaborn-v0_8")
```

```python
# Create some data
x = np.random.randn(10,4)
x
```

```python
# Create a DataFrame
df = pd.DataFrame(x, columns=["a", "b", "c", "d"])
df
```

```python
# Create plot using pyplot api
ax = df.plot(kind='bar')

# Add some labels and a title
ax.set(title="Random Number Bar Graph from DataFame",
       xlabel="Row number",
       ylabel="Random Number")

# Make the legend visible
ax.legend().set_visible(True)
```

```python
# Set the style
plt.style.use("seaborn-v0_8-whitegrid")

# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                    cmap="winter")

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# Customizing the y and x axis limitations

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                     cmap="winter")

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Level",
        ylabel="Cholesterol")

# Change the x-axis limits of ax0
ax0.set_xlim([50, 80])

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a meanline to the horizontal
ax0.axhline(over_50["chol"].mean(), linestyle="--");

# AGE AND CHOLESTEROL
# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                     cmap="winter")
# Customize ax1
ax1.set(title="Heart Disease: Age and Max Heart Rate(Thalach)",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Change the x and y axis limits of ax1
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target");

# Add mean line to the horizontal and vertical axes
ax1.axhline(over_50["thalach"].mean(), linestyle="--")
ax1.axvline(over_50["age"].mean(), linestyle="--", color="orange")

# Add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
