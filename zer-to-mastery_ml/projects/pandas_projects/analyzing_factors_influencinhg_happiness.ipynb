## Find the happiest and least happy countries.

```python
# Happiest Country
happiest_country = data[data["Score"] == data["Score"].max()]
print("Happiest Country:")
print(happiest_country)

# Least happy country
least_happy_country = data[data["Score"] == data["Score"].min()]
print("\nLeast Happy Country:")
print(least_happy_country)
```
