## Matplotlib example workflow

```python
# Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height)

# Plot data
ax.plot(x, y)

# Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# Save and show
fig.savefig("../matplotlib/sample_plot.png")
```

## Making Figures with NumPy arrays


```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make a scatter

fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Create another scatter
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="My Nut Butter Store",
       ylabel="Price ($)");
```

```python
# To present a horizontal bar
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

## Which one should be used? (pyplot or matplotlib OO method?)
1. When plotting something quickly, use pyplot method
2. When plotting something more advanced, use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(x="age", y="chol", kind="scatter", c="target"); # c="target" is the colour
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
# ax.set_xlim([45, 100]);
```
